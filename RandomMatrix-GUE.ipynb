```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
from RandomMatrix import *
np.random.seed(1)
```

Problem(b)

```python
N = 2
sample = 2000
multiS = multi_space(N,sample,'GUE')
meanS = np.sum(multiS,axis=0)/sample
print(meanS)
n_multiS = multiS/meanS
```

```python
f, ax1 = plt.subplots(1, 1, figsize=(10,6))
ax1.hist(n_multiS,bins=20)
ax1.set_title('n=1')
ax1.set_ylabel('Frequency')
ax1.set_xticks(np.linspace(n_multiS.min(),n_multiS.max(),11))
plt.savefig('plot/GUE_b_2_1.eps')
plt.show()
```

Problem(c)

```python
N = 2
sample = 20000
multiS = multi_space(N,sample,'GUE')
meanS = np.sum(multiS,axis=0)/sample
print('Estimated: %f, Analytical: %f'%(meanS, 4/np.sqrt(np.pi)))
n_multiS = multiS/meanS
```

```python
f, ax1 = plt.subplots(1, 1, figsize=(10,6))
ax1.hist(n_multiS,bins=20,density=1)
X = np.linspace(n_multiS.min(),n_multiS.max(),101)
ax1.plot(X,32/np.pi**2*X**2*np.exp(-X**2*4/np.pi) ,'--')
ax1.set_ylabel('Probability Density')
ax1.set_xticks(np.linspace(n_multiS.min(),n_multiS.max(),11))
plt.savefig('plot/GUE_c_2_1.eps')
plt.show()
```
