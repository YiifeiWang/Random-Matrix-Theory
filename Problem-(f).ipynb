```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
from RandomMatrix import *
np.random.seed(1)
```

Real Wigner matrix

```python
N1 = 400
H1 = Wigner_Real(N1)
eigs1=la.eigvalsh(H1)
N2 = 900
H2 = Wigner_Real(N2)
eigs2=la.eigvalsh(H2)
```

```python
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,6))
ax1.hist(eigs1,bins=20)
ax1.set_title('N=400')
ax1.set_xlabel('Eigenvalue')
ax1.set_ylabel('Frequency')
ax1.set_xticks(np.linspace(eigs1.min(),eigs1.max(),11))
ax2.hist(eigs2,bins=20)
ax2.set_title('N=900')
ax2.set_xlabel('Eigenvalue')
ax2.set_ylabel('Frequency')
ax2.set_xticks(np.linspace(eigs2.min(),eigs2.max(),11))
plt.savefig('plot/f_rw_1.eps')
plt.show()
```

```python
N1 = 1000
H1 = Wigner_Real(N1)
eigs1=la.eigvalsh(H1)
N2 = 2000
H2 = Wigner_Real(N2)
eigs2=la.eigvalsh(H2)
```

```python
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,6))
ax1.hist(eigs1/np.sqrt(N1),bins=40)
ax1.set_title('N=1000')
ax1.set_xlabel('Rescaled Eigenvalue')
ax1.set_ylabel('Frequency')
ax1.set_xticks(np.linspace(-2,2,11))
ax2.hist(eigs2/np.sqrt(N2),bins=40)
ax2.set_title('N=2000')
ax2.set_xlabel('Rescaled Eigenvalue')
ax2.set_ylabel('Frequency')
ax2.set_xticks(np.linspace(-2,2,11))
plt.savefig('plot/f_rw_2.eps')
plt.show()
```

```python
N3 = 3000
H3 = Wigner_Real(N3)
eigs3=la.eigvalsh(H3)
N4 = 4000
H4 = Wigner_Real(N4)
eigs4=la.eigvalsh(H4)
```

```python
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,6))
ax1.hist(eigs3/np.sqrt(N3),bins=40)
ax1.set_title('N=3000')
ax1.set_xlabel('Rescaled Eigenvalue')
ax1.set_ylabel('Frequency')
ax1.set_xticks(np.linspace(-2,2,11))
ax2.hist(eigs4/np.sqrt(N4),bins=40)
ax2.set_title('N=4000')
ax2.set_xlabel('Rescaled Eigenvalue')
ax2.set_ylabel('Frequency')
ax2.set_xticks(np.linspace(-2,2,11))
plt.savefig('plot/f_rw_3.eps')
plt.show()
```

```python
f, (ax1,ax2) = plt.subplots(1, 2, figsize=(20,6))
X = np.linspace(-2,2,201)
ax1.hist(eigs3/np.sqrt(N3),bins=40,density=1)
ax1.plot(X,np.sqrt(4-X**2)/2/np.pi,'--')
ax1.set_title('N=3000')
ax1.set_xlabel('Rescaled Eigenvalue')
ax1.set_ylabel('Probability Density')
ax1.set_xticks(np.linspace(-2,2,11))
ax2.hist(eigs4/np.sqrt(N4),bins=40,density=1)
ax2.plot(X,np.sqrt(4-X**2)/2/np.pi,'--')
ax2.set_title('N=4000')
ax2.set_xlabel('Rescaled Eigenvalue')
ax2.set_ylabel('Probability Density')
ax2.set_xticks(np.linspace(-2,2,11))
plt.savefig('plot/f_rw_4.eps')
plt.show()
```
